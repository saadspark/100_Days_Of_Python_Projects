```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
file_path = '/content/drive/MyDrive/Mechine_Learning_DataSets/canada_per_capita_income.csv'
df = pd.read_csv(file_path)
df.head()
```

```python


from matplotlib import pyplot as plt
df.plot(kind='scatter', x='year', y='per capita income (US$)',)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df[['year']], df['per capita income (US$)'])
```

```python
model.score(df[['year']], df['per capita income (US$)'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['per capita income (US$)'], label='Actual Data')
plt.plot(df['year'], model.predict(df[['year']]), color='red', label='Linear Regression Fit')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Canada Per Capita Income Over Time with Linear Regression Fit')
plt.grid(True)
plt.legend()
plt.show()
```

```python
years_to_predict = pd.DataFrame({'year': range(2020, 2030)})
predicted_income = model.predict(years_to_predict[['year']])

predictions_df = years_to_predict.copy()
predictions_df['predicted per capita income (US$)'] = predicted_income

print("Predictions for 2020 to 2025:")
display(predictions_df)
```

```python
# @title year vs predicted per capita income (US$)

from matplotlib import pyplot as plt
predictions_df.plot(kind='scatter', x='year', y='predicted per capita income (US$)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
